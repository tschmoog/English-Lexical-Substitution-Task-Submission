# Helper functions:

Insert brackets helper function

```python
def brackets(string_list):
    b1 = '{'
    b2 = '}'
    holder_list = [b1 + x + b2 for x in string_list]    
    return holder_list
```

Read the data from a string and create lists of items 

```python

#Read the data from a string and create lists of items
#List of word: context pairs that can be interated through
word_list = [] #List of words to be replaced
pos_list = [] #List of parts of speech for previous words
context_list = [] #Context of each word

for document in root:
    word_pos = document.attrib['item']
    word_pos_list = word_pos.split('.')
    word = word_pos_list[0]
    pos = word_pos_list[1:]
    
    for instance in document:
        for sentence in instance:
            #print 'Word:'
            #print word
            word_list.append(word)
            #print 'pos:'
            #print pos
            pos_list.append(pos)
            #print 'sentence:'
            #print sentence.text
            context_list.append(sentence.text)
        
    

```

Step 1: Find context for the word using the Lesk algorithims

```python
#Create lists of contexts using lesk algorithims

#Original lesk
org_lesk = []

for i in range(len(word_list)):
    x = lesk(context_list[i], word_list[i]).name()
    x = conv_to_plain(x)
    org_lesk.append(x)
    
print org_lesk
print '------------------------'

#Simple lesk
simple_lesk_list = []
for i in range(len(word_list)):
    x = simple_lesk(context_list[i], word_list[i]).name()
    x = conv_to_plain(x)
    simple_lesk_list.append(x)

    
print simple_lesk_list
print '------------------------'

#Adapted lesk 
adapt_lesk_list = []
for i in range(len(word_list)):
    x = adapted_lesk(context_list[i], word_list[i]).name()
    x = conv_to_plain(x)
    adapt_lesk_list.append(x)

print adapt_lesk_list
print '------------------------'

#Cosine Lesk
cosine_lesk_list = []
for i in range(len(word_list)):
    x = cosine_lesk(context_list[i], word_list[i]).name()
    x = conv_to_plain(x)
    cosine_lesk_list.append(x)

print cosine_lesk_list
print '------------------------'
    
```

Step 2: Generate alternative words. 


Find Synonyms using Lin theasaurus

```python
#Generate Synonyms using Lin thesaurus

lin_thes_list = []
counter = 0
for word in cosine_lesk_list: #adapt_lesk_list: #simple_lesk_list: #org_lesk: #

    if pos_list[counter][-1] == 'n':    
        lin_thes_list.append(thes.synonyms(word)[1][1])
    elif pos_list[counter][-1] == 'v':
        lin_thes_list.append(thes.synonyms(word)[2][1])
    else:
        lin_thes_list.append(thes.synonyms(word)[0][1])

        
    counter = counter + 1
    
print lin_thes_list



```

## Find Synonyms using WordNet

```python
wordnet_list = []
for word in simple_lesk_list: #org_lesk: #cosine_lesk_list: # adapt_lesk_list: #
    temp_list = []
    flattened_list = []
    for x in word.hypernyms():
        temp_list.append(x.lemma_names())
    temp_list.append(word.lemma_names())
    flattened_list = [y for x in temp_list for y in x]
    
    wordnet_list.append(flattened_list)

print wordnet_list
```

## Select the most apropiate response from the list of generated synonyms using word2vec

```python
counter = 0
similarity_list = []
for word in word_list:
    holder_list = []
    for synonym in lin_thes_list[counter]: #
        try:
            ss = mymodel.similarity(word,synonym)
            if ss < 0.98:
                holder_list.append(ss)
            elif ss > 0.98:
                holder_list.append(0.1)
        except:
            holder_list.append(0)

    
    similarity_list.append(holder_list)        
    counter += 1
    
print similarity_list[0]
```

Step 3: Evaluate against the gold standard

```python
def score_results(filename):
    pipe = subprocess.Popen(["perl", "score.pl",filename," gold.trial"], stdout=subprocess.PIPE)
    out, err = pipe.communicate()
    if err:
        print err

# Write full output to file
    output_file = codecs.open(filename + '.score', 'w', encoding='utf-8')
    output_file.write(out)
    output_file.close()

    return out

```
